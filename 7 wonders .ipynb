```python
symbol_to_number = dict()
symbol_to_number['hammer'] = 13
symbol_to_number['drop'] = 14
symbol_to_number['star'] = 15
symbol_to_number['mask'] = 16
symbol_to_number['camel'] = 17
symbol_to_number['shop'] = 18
symbol_to_number['horseshoe'] = 19
symbol_to_number['stupa'] = 20
symbol_to_number['goal'] = 21
symbol_to_number['lamp'] = 22
symbol_to_number['balance'] = 23
symbol_to_number['book'] = 24


max_symb_val = max([symbol_to_number[key] for key in symbol_to_number])
```

```python
amount_of_resource = 28

deck_of_cards = dict()
deck_of_cards['phase 1'] = dict()


for i, base_resource_name in enumerate(['Лесопилка', 'Каменный карьер', 
                                        'Глиняный карьер', 'Рудник',
                                        'Стекольный завод', 
                                        'Печатный станок',
                                        'Ткацкий станок']):
    for j in [1, 2]:
        card_name = '{} - {}'.format(base_resource_name, j)
        deck_of_cards['phase 1'][card_name] = dict()
        if i < 4:
            deck_of_cards['phase 1'][card_name]['color'] = 'brown'
        else:
            deck_of_cards['phase 1'][card_name]['color'] = 'grey'
        deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
        deck_of_cards['phase 1'][card_name]['resources'][i] = 1.0
        deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
        deck_of_cards['phase 1'][card_name]['in'] = []
        deck_of_cards['phase 1'][card_name]['out'] = []
        
        
        
for card_name, resources in zip(['Лесная Ферма', 'Раскопки', 'Глиняная Яма',
                                 'Лесоповал', 'Лесная Пещера', 'Шахта'],
                                [[0, 2], [1, 2], [2, 3], [0, 1], [0, 3],
                                 [1, 3]]):
    deck_of_cards['phase 1'][card_name] = dict()
    deck_of_cards['phase 1'][card_name]['color'] = 'brown'
    deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
    deck_of_cards['phase 1'][card_name]['resources'][resources[0]] = 1.0
    deck_of_cards['phase 1'][card_name]['resources'][resources[1]] = 1.0
    deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
    deck_of_cards['phase 1'][card_name]['cost'][7] = 1.0
    deck_of_cards['phase 1'][card_name]['in'] = []
    deck_of_cards['phase 1'][card_name]['out'] = []
    
    
    
for k, (basic_card_name, symbol) in enumerate(zip(['Ломбард', 'Бани', 'Алтарь', 'Театр'],
                                                  ['hammer', 'drop', 'star', 'mask'])):
    for j in [1, 2]:
        card_name = '{} - {}'.format(basic_card_name, j)
        deck_of_cards['phase 1'][card_name] = dict()
        deck_of_cards['phase 1'][card_name]['color'] = 'blue'
        deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
        deck_of_cards['phase 1'][card_name]['resources'][8] = 3.0
        deck_of_cards['phase 1'][card_name]['resources'][symbol_to_number[symbol]] = 1.0
        deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
        if basic_card_name == 'Бани':
            deck_of_cards['phase 1'][card_name]['cost'][1] = 1.0
        deck_of_cards['phase 1'][card_name]['in'] = []
        deck_of_cards['phase 1'][card_name]['out'] = [symbol]
        
        
        
for k, (basic_card_name, symbol) in enumerate(zip(['Таверна', 'Рынок', 'Западный Торговый Пост',
                                                    'Восточный Торговый пост'],
                                                   [np.nan, 'camel', 'shop', 'shop'])):
    for j in [1, 2]:
        card_name = '{} - {}'.format(basic_card_name, j)
        deck_of_cards['phase 1'][card_name] = dict()
        deck_of_cards['phase 1'][card_name]['color'] = 'yellow'
        deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
        if basic_card_name == 'Таверна':
            deck_of_cards['phase 1'][card_name]['resources'][7] = 5.0
        else:
            deck_of_cards['phase 1'][card_name]['resources'][k + max_symb_val - 1] = 1.0
            deck_of_cards['phase 1'][card_name]['resources'][symbol_to_number[symbol]] = 1.0
        deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
        deck_of_cards['phase 1'][card_name]['in'] = []
        if basic_card_name != 'Таверна': 
            deck_of_cards['phase 1'][card_name]['out'] = [symbol]
            
            
            
            
for basic_card_name, cost in zip(['Частокол', 'Бараки', 'Сторожевая Башня'],
                                 [0, 3, 2]):
    for j in [1, 2]:
        card_name = '{} - {}'.format(basic_card_name, j)
        deck_of_cards['phase 1'][card_name] = dict()
        deck_of_cards['phase 1'][card_name]['color'] = 'red'
        deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
        deck_of_cards['phase 1'][card_name]['resources'][9] = 1.0
        deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
        deck_of_cards['phase 1'][card_name]['cost'][cost] = 1.0
        deck_of_cards['phase 1'][card_name]['in'] = []
        deck_of_cards['phase 1'][card_name]['out'] = []
        
        
        
for k, (basic_card_name, cost, symbols) in enumerate(zip(['Аптека', 'Мастерская', 'Скрипторий'],
                                                     [6, 4, 5], [['horseshoe', 'stupa'],
                                                                 ['goal', 'lamp'],
                                                                 ['balance', 'book']])):
    for j in [1, 2]:
        card_name = '{} - {}'.format(basic_card_name, j)
        deck_of_cards['phase 1'][card_name] = dict()
        deck_of_cards['phase 1'][card_name]['color'] = 'green'
        deck_of_cards['phase 1'][card_name]['resources'] = [0.0 for i in range(amount_of_resource)]
        deck_of_cards['phase 1'][card_name]['resources'][10 + k] = 1.0
        deck_of_cards['phase 1'][card_name]['cost'] = [0.0 for i in range(8)]
        deck_of_cards['phase 1'][card_name]['cost'][cost] = 1.0
        deck_of_cards['phase 1'][card_name]['in'] = []
        deck_of_cards['phase 1'][card_name]['out'] = [symbol for symbol in symbols]
        for symbol in symbols:
            deck_of_cards['phase 1'][card_name]['resources'][symbol_to_number[symbol]] = 1.0
        
        
        
for name in ['Стекольный завод - 2', 'Печатный станок - 2',
             'Ткацкий станок - 2', 'Лесная Ферма', 'Шахта',
             'Бани - 2', 'Ломбард - 2', 'Театр - 2',
             'Рынок - 2', 'Западный Торговый Пост - 2', 
             'Восточный Торговый пост - 2', 'Частокол - 2',
             'Мастерская - 2']:
    del deck_of_cards['phase 1'][name]
    
    
# сделать символы в визе полей
```
